```python
import nltk
from nltk.tokenize import word_tokenize
import string
import matplotlib.pyplot as plt

# Downloading required packages
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

# Sentences dataset
sentences = [
    "I absolutely love studying cybersecurity, it’s so exciting!",
    "This online course was amazing and really helped me understand the topic.",
    "I’m so happy with my new laptop for studying and work.",
    "The teacher was great and explained everything clearly.",
    "What a beautiful study setup, I feel so focused!",
    "I feel motivated and excited to start my new project.",
    "The customer support was friendly and solved my issue fast.",
    "Buying this keyboard was a good decision, it feels awesome.",
    "The app design is amazing, very clean and modern.",
    "I had a good experience with this online store.",
    "This is the worst course I’ve ever taken.",
    "I feel sad and disappointed with my test results.",
    "The delivery was terrible and took forever to arrive.",
    "I hated how slow my laptop became after the update.",
    "The teacher’s explanation was awful and confusing.",
    "This product is bad quality and not worth the price.",
    "I regret buying this phone, it’s horrible.",
    "The website looks cheap and unprofessional.",
    "I’m angry that my payment failed again.",
    "It was a horrible experience trying to contact their support.",
    "The study material was good but could be more detailed.",
    "I’m not sure if I made the right decision choosing this course.",
    "The headphones sound great and feel comfortable.",
    "I love how clean and fast this app works.",
    "The instructor was awesome and made learning fun.",
    "The service was disappointing and not what I expected.",
    "The session was okay, not too exciting but fine overall.",
    "I’m happy I passed the exam, it feels great!",
    "The new update made the app terrible and full of bugs.",
    "What an amazing improvement after the last version!"
]

# Positive words list
positiveWords = [
    "love", "amazing", "happy", "excited", "great", "beautiful", "motivated",
    "friendly", "awesome", "good", "clean", "modern", "comfortable", "fast",
    "fun", "improvement", "perfect", "helpful", "focused", "worth"
]

# Negative words list
negativeWords = [
    "worst", "sad", "disappointed", "terrible", "hated", "awful", "bad",
    "cheap", "horrible", "angry", "slow", "regret", "confusing", "unprofessional",
    "failed", "disappointing", "boring", "buggy", "broken", "frustrating"
]

# Finding what sentences are positive, negative, or neutral
results = []
for sentence in sentences:
    text = sentence.lower().translate(str.maketrans("", "", string.punctuation))
    words = word_tokenize(text)
    posWord = sum(1 for w in words if w in positiveWords)
    negWord = sum(1 for w in words if w in negativeWords)

    if posWord > negWord:
        sentiment = "positive"
    elif negWord > posWord:
        sentiment = "negative"
    else:
        sentiment = "neutral"

    results.append((sentence, sentiment))

# Counting number of positive, negative, or neutral sentences
sentimentCounts = {"positive": 0, "negative": 0, "neutral": 0}
for _, label in results:
    sentimentCounts[label] += 1

# Drawing graph
plt.figure(figsize=(7, 5))
plt.bar(sentimentCounts.keys(), sentimentCounts.values(), color=['green', 'red', 'gray'])
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Number of Sentences")
plt.show()

# Printing results
for sentence, sentiment in results:
    print(f"Sentence: {sentence}  =>  Sentiment: {sentiment}")

print("____________End____________")
```
